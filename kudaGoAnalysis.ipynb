```python
import pandas as pd
from geopy.distance import distance
```

```python
df = pd.read_json('Places_spb_list.json')
df.head()
```

```python
pd.DataFrame(df.coords.tolist())
```

```python
# может это и не надо выносить в отдельные столбцы
df.join(pd.DataFrame(df.coords.tolist()))
```

```python
# места, у которых нет ближайшего метро
df[df['subway'] == '']
```

```python
# Оставляем только одну станцию метро
df['subway'] = list(map(lambda x: x.split(',')[0].casefold(), df['subway']))
# делаем все названия одного регистра, чтобы было удобнее
df['title'] = df['title'].apply(lambda x: x.casefold())
df.head()
```

```python
def distance_to_subway(index):
    try:
        return distance(df.loc[index].coords.values(), 
                     df[df['title'] == f'станция метро «{df.loc[index].subway}»'].iloc[0].coords.values()).km
    except IndexError:
        return None
```

```python
df['distance_to_subway, km'] = list(map(lambda i: distance_to_subway(i), df.index))
```

```python
df.head()
```

```python
# расстояние то центра города (координаты цетра города в этом случае, например: Казанский собор)
kazan_cathedral = df[df['title']=='казанский собор'].iloc[0].coords.values()
def distance_to_center(index):
    try:
        return distance(df.loc[index].coords.values(), kazan_cathedral).km
    except IndexError:
        return None
```

```python
df['distance_to_center, km'] = list(map(lambda i: distance_to_center(i), df.index))
```

```python
df.head()
```

```python
# кол-во сооружений таких же категорий в радиусе 1 км
df['categories'] = df['categories'].apply(', '.join)
```

```python
df.head()
```

```python
def count_similar_places(df):
    l = []
    for j in range(len(df)):
        categor = df.loc[j].categories
        suit = df['categories'] == categor
        suit_indexes = df.index[suit].to_list()
        count = 0
        for i in suit_indexes:
            dist = distance(df.loc[j].coords.values(), df.loc[i].coords.values()).km
            if dist > 0 and dist < 1:
                count += 1
        l.append(count)
    return l
```

```python
df['similar_places'] = count_similar_places(df)
```
